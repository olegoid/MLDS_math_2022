## Задание 1

Найдите экстремумы следующих функций (вручную, где возможно) и укажите их тип (максимум, минимум):

a) $f(x) = x^2 - 30x + 1$

Найдем произвоидную:

$f'(x) = 2(x - 15)$

Дальше ищем критические точки функции:

$2(x - 15) = 0$

$x = 15$ - *это минимум*

b) $g(x) = -x^3 + 5x - 7x^2 + 17$

Найдем произвоидную:

$f'(x) = 5 - 14x - 3x^2$

Дальше ищем критические точки функции:

$5 - 14x - 3x^2 = 0$

$(x + 5)(3x- 1) = 0$

$x = -5$ - *это минимум*, $x = 1 / 3$ - *это максимум*

c) $h(x) = x^2 - 1/x$

Найдем произвоидную:

$f'(x) = 2x + 1/x^2$

Дальше ищем критические точки функции:

$2x + 1/x^2 = 0$

$x = - 1/\sqrt[3]{2}$ 

$f'(x)$ определена определена для всех $х$ где $x \neq 0$

Разделим область определиния на два интервала относитиельно точки:
$(-\infty, - 1/\sqrt[3]{2})$, $(- 1/\sqrt[3]{2}, 0)$

и выберем тестовые точки - $x = -1$, $x = -1/2$

| $x$         | $f'(x)$     | sign |
| ----------- | ----------- |------|
| -1          | -1          |   _  |
| -1/2        | 3           |   +  |

Точка является *локальным минимумом* (тк $f'(x)$ меняет знак с - на +)

d) $j(x) = 1/x^5 + x^{17} + 8$

Найдем произвоидную:

$f'(x) = -(5 - 17 x^22)/x^6$

Дальше ищем критические точки функции:

$-(5 - 17 x^22)/x^6 = 0$

$x = -\sqrt[22]{(5/17)} \approxeq -0.94589$ - точка максимума, $x = \sqrt[22]{(5/17)} \approxeq 0.94589$ - точка минимума.

## Задание 6

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
def gradient_descent(f, x_start, learning_rate, epsilon, num_iterations, delta):
    x_curr = x_start
    df_x = sp.diff(f(x)) # производная

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):

      x_new = x_curr - learning_rate * df_x.subs(x, x_curr) 
      trace.append(x_new)

      if abs(x_new - x_curr) < epsilon:
        break

      if abs(df_x.subs(x, x_curr)) < delta:
        break

      x_curr = x_new

    return x_curr, trace
```

## Задание 8

Реализуйте [метод Ньютона](https://translated.turbopages.org/proxy_u/en-ru.ru.c97ef85d-631c4b6a-ff938028-74722d776562/https/en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
def newton_descent(f, x_start, epsilon, num_iterations):
    x_curr = x_start
    df_x = sp.diff(f(x))
    d2f_x = sp.diff(df_x)

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):

      x_new = x_curr - df_x.subs(x, x_curr) / d2f_x.subs(x, x_curr)
      trace.append(x_new)

      if abs(x_new - x_curr) < epsilon:
        break

      x_curr = x_new

    return x_curr, trace
```

## Задание 11*

Реализуйте оптимизацию метода градиентного спуска - [Momentum](https://ml-handbook.ru/chapters/optimization/intro#метод-инерции-momentum).

```python
def momentum_descent(f, x_start, learning_rate, epsilon, num_iterations, beta):
    x_curr = x_start
    x_prev = x_curr
    df_x = sp.diff(f(x)) # производная

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):

      x_new = x_curr - learning_rate * df_x.subs(x, x_curr)  + beta * (x_curr - x_prev)
      trace.append(x_new)

      if abs(x_new - x_curr) < epsilon:
        break
    
      x_prev = x_curr
      x_curr = x_new

    return x_curr, trace
```
