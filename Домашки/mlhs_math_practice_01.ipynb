## Задание 1

Найдите экстремумы следующих функций (вручную, где возможно) и укажите их тип (максимум, минимум):

a) $f(x) = x^2 - 30x + 1$

Найдем произвоидную:

$f'(x) = 2(x - 15)$

Дальше ищем критические точки функции:

$2(x - 15) = 0$

$x = 15$ - *это минимум*

b) $g(x) = -x^3 + 5x - 7x^2 + 17$

Найдем произвоидную:

$f'(x) = 5 - 14x - 3x^2$

Дальше ищем критические точки функции:

$5 - 14x - 3x^2 = 0$

$(x + 5)(3x- 1) = 0$

$x = -5$ - *это минимум*, $x = 1 / 3$ - *это максимум*

c) $h(x) = x^2 - 1/x$

Найдем произвоидную:

$f'(x) = 2x + 1/x^2$

Дальше ищем критические точки функции:

$2x + 1/x^2 = 0$

$x = - 1/\sqrt[3]{2}$ 

$f'(x)$ определена определена для всех $х$ где $x \neq 0$

Разделим область определиния на два интервала относитиельно точки:
$(-\infty, - 1/\sqrt[3]{2})$, $(- 1/\sqrt[3]{2}, 0)$

и выберем тестовые точки - $x = -1$, $x = -1/2$

| $x$         | $f'(x)$     | sign |
| ----------- | ----------- |------|
| -1          | -1          |   _  |
| -1/2        | 3           |   +  |

Точка является *локальным минимумом* (тк $f'(x)$ меняет знак с - на +)

d) $j(x) = 1/x^5 + x^{17} + 8$

Найдем произвоидную:

$f'(x) = -(5 - 17 x^22)/x^6$

Дальше ищем критические точки функции:

$-(5 - 17 x^22)/x^6 = 0$

$x = -\sqrt[22]{(5/17)} \approxeq -0.94589$ - точка максимума, $x = \sqrt[22]{(5/17)} \approxeq 0.94589$ - точка минимума.

## Задание 2

Найдите экстремумы этих функций в Python.

a) $f(x) = x^2 - 30x + 1$

```python
import sympy as sp

def f1(x):
  return x**2 - 30*x + 1

x = sp.Symbol('x')

df1_x = sp.diff(f1(x))
print(df1_x)
print(sp.solve(df1_x, x))
```

b) $g(x) = -x^3 + 5x - 7x^2 + 17$

```python
def f2(x):
  return -x ** 3 + 5 * x - 7 * x ** 2 + 17

df2_x = sp.diff(f2(x))
print(df2_x)
print(sp.solve(df2_x, x))
```

c) $h(x) = x^2 - 1/x$

```python
def f3(x):
  return x ** 2 - 1 / x

df3_x = sp.diff(f3(x))
print(df3_x)
print(sp.solve(df3_x, x))
```

d) $j(x) = 1/x^5 + x^{17} + 8$

```python
def f4(x):
  return 1 / (x ** 5) + x ** 17 + 8

df4_x = sp.diff(f4(x))
print(df4_x)
print(sp.solve(df4_x, x))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

x_values = [x for x in np.arange(-30, 60, 0.1)]
f_values = [f1(x) for x in x_values]

plt.figure(figsize=(10,10))
plt.axvline(x=0, c = 'black')
plt.axhline(y=0, c = 'black')

plt.scatter([15],[f1(15)],color=['red'])

plt.plot(x_values, f_values)

plt.title('График функции $f(x) = x^2 - 30x + 1$ c экстремумами')

plt.xlabel('x')
plt.ylabel('f(x)')
```
