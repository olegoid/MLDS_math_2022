Собственные значения и собственные векторы оператора

```python
A = np.array([[-1,-6],[2,6]])
w, v = np.linalg.eig(A)

print('Первое собственное число и вектор:', w[0], v[:,0])
print('Второе собственное число и вектор:', w[1], v[:,1])
```

Нарисовать преобразование пространства (куда переходят базисные векторы), заданное оператором. 

```python
import matplotlib.pyplot as plt

origin = np.array([[0, 0],[0, 0]]) # origin point

fig, ax = plt.subplots(nrows=1, ncols=2)

U = np.array([[1,0], [0,1]])
ax[0].quiver(*origin, U[:,0], U[:,1], color='g',scale=7)
ax[0].set_xlim([-1,1])
ax[0].set_ylim([-1,1])

V = np.array([[3,1], [2,2]])
ax[1].quiver(*origin, V[:,0], V[:,1], color='b',scale=7)
ax[1].set_xlim([-1,1])
ax[1].set_ylim([-1,1])

plt.show()
```
